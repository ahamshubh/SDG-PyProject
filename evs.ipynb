```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/shuvish/Desktop/data.csv')
sns.countplot(x='VehicleEnergySource', data=df,palette=['#FF6347', '#4682B4', '#32CD32'])
plt.title('Type of fuel used for vehicles')
plt.xlabel('Energy Type')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('C:/Users/shuvish/Desktop/data.csv')
counts = df['SustainableBrands'].value_counts()
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots()
ax.pie(counts, labels=['True', 'False'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2", 2))
ax.axis('equal')
ax.set_title("Use of Sustainable Brands")
plt.show()
```

More than 54% of people have agreed on using sustainable brands in daily lives. this indicates a resposnible behaviour backed by awarness about environment's sustainability. SDG12(Responsible Consumption and Production) is accomplished at a good level

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('C:/Users/shuvish/Desktop/data.csv')
category_counts = df['DisposalMethods'].value_counts()
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(8, 6))  
ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, 
       colors=sns.color_palette("Set2", 4))  
ax.axis('equal')
ax.set_title('Disposal Methods')
plt.show()
```

Trends show that majority of the waste is disposed by landfilling, but harmful substances and plastics put SDG13 at stake while other remaining methods strengthen it and also help in acomplishing SDG15.

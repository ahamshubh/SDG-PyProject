the dataset in consideration is a compilation of all the responses gathered from a housing society during a survey. Survey link is present in the repository.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from google.colab import files
uploaded = files.upload()


# Set aesthetic Seaborn theme
sns.set(style='whitegrid', palette='muted', font_scale=1.2)

# Load dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='teal')
plt.title("Age Distribution of Participants")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='VehicleEnergySource', palette='Set2')
plt.title("Vehicle Energy Source Preferences")
plt.xlabel("Energy Source")
plt.ylabel("Number of People")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='HomeType', palette='pastel')
plt.title("Distribution of Home Types")
plt.xlabel("Home Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['SustainableBrands'].value_counts().plot.pie(
    autopct='%1.1f%%', startangle=90,
    colors=['lightgreen', 'salmon']
)
plt.title("Usage of Sustainable Brands")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='EnvironmentalAwareness', palette='viridis')
plt.title("Environmental Awareness Levels")
plt.xlabel("Awareness Level (1-5)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['MonthlyWaterConsumption'], color='skyblue')
plt.title("Monthly Water Consumption")
plt.xlabel("Water Consumption (Liters)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
order = ["Never", "Rarely", "Sometimes", "Often"]
sns.countplot(data=df, x='UsingPlasticProducts', order=order, palette='coolwarm')
plt.title("Plastic Product Usage Frequency")
plt.xlabel("Frequency")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='DisposalMethods', palette='Accent')
plt.title("Preferred Waste Disposal Methods")
plt.xlabel("Disposal Method")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='EnvironmentalAwareness', y='MonthlyWaterConsumption', hue='SustainableBrands', style='SustainableBrands', palette='Set1')
plt.title("Water Consumption vs Environmental Awareness")
plt.xlabel("Environmental Awareness Level")
plt.ylabel("Monthly Water Consumption (L)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='HomeType', y='MonthlyWaterConsumption', palette='pastel')
plt.title("Monthly Water Consumption by Home Type")
plt.xlabel("Home Type")
plt.ylabel("Water Consumption (L)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='VehicleEnergySource', y='EnvironmentalAwareness', palette='Set2')
plt.title("Environmental Awareness by Vehicle Energy Source")
plt.xlabel("Vehicle Energy Source")
plt.ylabel("Awareness Level")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
order = ["Never", "Rarely", "Sometimes", "Often"]
sns.boxplot(data=df, x='UsingPlasticProducts', y='MonthlyWaterConsumption', order=order, palette='coolwarm')
plt.title("Water Consumption by Plastic Usage Frequency")
plt.xlabel("Plastic Usage Frequency")
plt.ylabel("Monthly Water Consumption (L)")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue="SustainableBrands", vars=["Age", "EnvironmentalAwareness", "MonthlyWaterConsumption"], palette='husl')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```
